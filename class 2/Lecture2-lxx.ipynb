```python
def get_probability_from_counts(count):
    all_occurances = sum(count.values())
    def get_prob(item):
        return count[item] / all_occurances
    return get_prob

get_char_prob = get_probability_from_counts(all_character_counters)

def get_char_probability(char):
    all_occurances = sum(all_character_counters.values())
    return all_character_counters[char] / all_occurances
```

```python
def prob_of_string(string):
    return reduce(mul, [get_char_prob(c) for c in string])
```

```python
def get_probability_performance(language_model_func, pair):
    for (p1,p2) in pairs:
        print('*'*18)
        print('\t\t {} with probability {}'.format(p1, language_model_func(tokensize(p1))))
        print('\t\t {} with probability {}'.format(p2, language_model_func(tokensize(p2))))
```

```python
gram_length = 2
two_gram_counts = Counter(ALL_CHARACTER[i:i+gram_length] for i in range(len(ALL_CHARACTER) - gram_length))
```

```python
two_gram_counts.most_common()[:20]
```

```python
get_pair_prob = get_probability_from_counts(two_gram_counts)
```

```python
def get_2_gram_prob(word, prev):
    if get_pair_prob(word+prev) > 0:
        return get_pair_prob(word + prev) / get_char_prob(prev)
    else:
        return get_char_prob(word)
    
def get_2_gram_string_prob(string):
    probabilities = []
    for i,c in enumerate(string):
        prev = '<s>' if i == 0 else string[i-1]
        probabilities.append(get_2_gram_prob(c, prev))
    return reduce(mul, probabilities)
```

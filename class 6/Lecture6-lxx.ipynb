```python
import math
```

### inverse document frequency：逆向文件频率，总文件数 / 单词出现的文件数，取对数

```python
def inverse_document_frequency(word):
    eps = 1e-5
    return math.log10(len(all_occurances) / (sum(1 for w in all_occurances if word in w) + eps))
```

### Document Frequency:文件频率，在所有语料库中出现的次数，越大表示重要性越低，例如，“的”，“我们”

```python
def document_frequency(word):
    return sum(1 for w in all_occurances if word in w) / len(all_occurances)
```

```python
document_frequency('我们')
```

### TF--Term Frequence：词频，表示在一篇文章中出现的次数

```python
def tf(word, cut_word_counter):
    return cut_word_counter[word] / sum(cut_word_counter.values())
```

```python
tf('小米', Counter(content_of_xiaomi))
```

## TF-IDF是一种统计方法，用以评估一字词对于一个文件集或一个语料库中的其中一份文件的重要程度。字词的重要性随着它在文件中出现的次数成正比增加，但同时会随着它在语料库中出现的频率成反比下降。

```python
def tfidf(word, cut_words_counter):
    w_tf = tf(word, cut_words_counter)
    idf = inverse_document_frequency(word)
    return w_tf * idf
```

```python
tfidf('小米', Counter(content_of_xiaomi))
```

```python
tfidf('的', Counter(content_of_xiaomi))
```

```python
def get_words_importance(cut_words):
    importance = {
        w:tfidf(w,Counter(cut_words)) for w in set(cut_words)
    }
    return sorted(importance.items(), key = lambda x:x[1], reverse=True)
```

## Fine-tune

```python
from collections import defaultdict
```

```python
def get_related_words(initial_words, model):
    max_size = 2000
    seen = defaultdict(int)
    unseen = [initial_words]
    
    while unseen and len(seen) < max_size:
        node = unseen.pop(0)
        
#         if node in seen: continue
            
        new_expanding = [w for w, s in model.most_similar(node, topn=20)]
        unseen += new_expanding
        
        seen[node] += 1
        
#         for n in new_expanding:
#             seen[n] += 1;
            
    return seen
```

```python
related_words = get_related_words('中国', new_word2vec)
```
